```python
from sqlalchemy import text
```

```python
from sqlalchemy import create_engine
engine = create_engine("mysql+pymysql://root:test@127.0.0.1:3306/test?charset=utf8mb4", pool_size=5, max_overflow=300)
```

```python
with engine.connect() as conn:
    result = conn.execute(text("select 'hello world'"))
    print(result.all())
```
